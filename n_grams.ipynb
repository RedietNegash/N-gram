pre-processing the text

```python

"""
pre-processing the text
"""

def process_text(text):
    text = text.lower()
    text = text.replace(',', ' ')
    text = text.replace('/', ' ')
    text = text.replace('(', ' ')
    text = text.replace(')', ' ')
    text = text.replace('.', ' ')
    return text.split()
```

```python
"""
-> for the given words_list and number of consecutive words extract n consecutive words.

-> joing the strings

"""

def generate_ngrams(words_list, n):
    ngrams_list = []

    for num in range(0, len(words_list)):
        ngram = ' '.join(words_list[num:num + n])
        ngrams_list.append(ngram)

    return ngrams_list
```

```python
"""
-> process the input text and generate n-grams

"""


def process_and_generate_ngrams(text):

    words_list = process_text(text)
    bigrams = generate_ngrams(words_list, 2)

    trigrams = generate_ngrams(words_list, 3)
    print(trigrams)
    return trigrams
```

```python


"""
-> load the data from csv file and extract 'headline-text' row since it contains the document list texts
-> pass the text content to process and generate function
"""

import csv
file_path = '/content/abcnews-date-text.csv'
def process_csv(file_path):
    with open(file_path, mode='r', newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
          text=row['headline_text']
          process_and_generate_ngrams(text)

process_csv(file_path)
```
